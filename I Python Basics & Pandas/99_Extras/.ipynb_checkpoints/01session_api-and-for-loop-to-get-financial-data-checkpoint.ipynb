### Get Stock Price for `AAPL`

```python
pepa = 'AAPL'
```

```python
clave = 'idk'
```

```python
url
```

```python
for pepa in ['IBM', 'AAPL', 'TSLA']:

    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={pepa}&interval=5min&apikey={clave}'

    r = requests.get(url)

    dic = r.json()

    df = pd.DataFrame(dic['Time Series (5min)'])

    df = df.transpose()

    df['ticker'] = pepa

    df
```

```python
lista = []
```

```python
for pepa in ['IBM', 'AAPL', 'TSLA']:

    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={pepa}&interval=5min&apikey={clave}'

    r = requests.get(url)

    dic = r.json()

    df = pd.DataFrame(dic['Time Series (5min)'])

    df = df.transpose()

    df['ticker'] = pepa

    lista.append(df)
```

```python
pd.concat(lista)
```
